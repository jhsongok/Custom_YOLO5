```python
import torch
import cv2
import numpy as np
import time
import io
import base64
from IPython.display import HTML
```

```python
# Detection ÌïòÍ∏∞ Ï†ÑÏóê ÏõêÎ≥∏ ÎèôÏòÅÏÉÅÏùÑ Display
video = io.open('/content/gdrive/MyDrive/Custom_YOLO5/custom_fruit.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video width="30%" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4"/>
             </video>'''.format(encoded.decode('ascii')))
```

```python
file_name = '/content/gdrive/MyDrive/Custom_YOLO5/custom_fruit.mp4'  # Detection Ìï† ÏõêÎ≥∏ ÎèôÏòÅÏÉÅ
output_name = 'yolo5_output_video.mp4'  # Detection Îêú output ÎèôÏòÅÏÉÅ 
elapsed_time = 0       # Ï¥ù Í≤ΩÍ≥ºÏãúÍ∞Ñ Ï¥àÍ∏∞Ìôî 
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/gdrive/MyDrive/Custom_YOLO5/best.pt', force_reload=True)
```

```python
# ÏõêÎ≥∏ ÎèôÏòÅÏÉÅÏóêÏÑú video streamÏùÑ ÏùΩÏñ¥Ïò®Îã§.
cap = cv2.VideoCapture(file_name)
writer = None
if not cap.isOpened:
    print('--(!)Error opening video capture')
    exit(0)
while True:
    ret, frame = cap.read()
    if frame is None:
        # close the video file pointers
        cap.release()
        # close the writer point
        writer.release()
        print('--(!) No captured frame -- Break!')
        print("elapsed time {:.3f} seconds".format(elapsed_time))
        break
    
    start_time = time.time()

    results = model(frame)

    frame = np.squeeze(results.render())

    if writer is None and output_name is not None:
        fourcc = cv2.VideoWriter_fourcc(*"DIVX")
        writer = cv2.VideoWriter(output_name, fourcc, 20,
                (frame.shape[1], frame.shape[0]), True)
        
    # disk Ïóê frame ÏùÑ write Ìï©ÎãàÎã§.
    if writer is not None:
        writer.write(frame)
    
    process_time = time.time() - start_time
    elapsed_time += process_time   # Ï¥ù Í≤ΩÍ≥ºÏãúÍ∞Ñ ÎàÑÏ†Å
    print("=== A frame took {:.3f} seconds".format(process_time))
```
